```python
from pathlib import Path
import re
import pandas as pd
from docx import Document

# Input file
input_file = Path("/Users/liuzimo/Downloads/08Apr2026 quadruple test6-5.docx")

# Output file in same directory with same name stem
output_file = input_file.with_suffix(".csv")

# Read the Word document
doc = Document(input_file)
text = "\n".join(p.text for p in doc.paragraphs)

# More flexible regex: allows zero or more spaces between fields
pattern = re.compile(
    r"\[(?P<Timestamp>\d{2}[A-Za-z]{3}\d{4}\s+\d{2}:\d{2}:\d{2}\.\d{3})\].*?"
    r"Population:\s*(?P<Population>\d+)\s*"
    r"Susceptible:\s*(?P<Susceptible>\d+)\s*"
    r"Exposed:\s*(?P<Exposed>\d+)\s*"
    r"Infected:\s*(?P<Infected>\d+)\s*"
    r"Recovered:\s*(?P<Recovered>\d+)\s*"
    r"Dead:\s*(?P<Dead>\d*)"
)

rows = []

for line in text.splitlines():
    match = pattern.search(line)
    if match:
        row = match.groupdict()
        row["Population"] = int(row["Population"])
        row["Susceptible"] = int(row["Susceptible"])
        row["Exposed"] = int(row["Exposed"])
        row["Infected"] = int(row["Infected"])
        row["Recovered"] = int(row["Recovered"])
        row["Dead"] = int(row["Dead"]) if row["Dead"] else 0
        rows.append(row)

df = pd.DataFrame(rows, columns=[
    "Timestamp",
    "Population",
    "Susceptible",
    "Exposed",
    "Infected",
    "Recovered",
    "Dead"
])

df.to_csv(output_file, index=False)

print(f"Saved to: {output_file}")
print(f"Rows extracted: {len(df)}")
print(df.head().to_csv(index=False))
```
