```python
%matplotlib inline

import warnings
from itertools import combinations
from pathlib import Path
from decimal import Decimal, ROUND_HALF_UP, InvalidOperation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from scipy.integrate import odeint
from scipy.optimize import least_squares
from scipy.stats import mannwhitneyu, kruskal
from matplotlib.lines import Line2D

# Optional notebook-friendly display
try:
    from IPython.display import display, Markdown
    HAS_IPYTHON_DISPLAY = True
except ImportError:
    HAS_IPYTHON_DISPLAY = False
    def display(x):
        print(x)
    def Markdown(x):
        return x


# =========================================================
# USER SETTINGS
# =========================================================

CITY_NAME = "Glasgow"

RUN_END = 900
COMMON_GRID = np.arange(0, RUN_END + 1, 1, dtype=float)

OUTPUT_DIR = Path("/Users/liuzimo/Desktop/MC data/ODE_workflow_outputs_glasgow")
CSV_OUT = OUTPUT_DIR / "csv"
PLOT_OUT = OUTPUT_DIR / "plots"

CSV_OUT.mkdir(parents=True, exist_ok=True)
PLOT_OUT.mkdir(parents=True, exist_ok=True)

# Show figures in Jupyter Notebook output area
SHOW_PLOT_PREVIEW = True

# CSV export rule for most files:
# keep digits up to the 2nd non-zero digit after the decimal point
CSV_NONZERO_DECIMALS = 2

ROUND_LABELS = {
    "round1": "Round 1",
    "round2": "Round 2",
    "round3": "Round 3",
    "round4": "Round 4",
}

# ---------------------------------------------------------
# FONT SETTINGS
# ---------------------------------------------------------
TITLE_FS = 20
AXIS_LABEL_FS = 18
TICK_FS = 14
LEGEND_FS = 15
SUPTITLE_FS = 18

# Smaller fonts specifically for crowded boxplots
BOXPLOT_TITLE_FS = 14
BOXPLOT_XTICK_FS = 10
BOXPLOT_YTICK_FS = 12

# Outside legend font
OUTSIDE_LEGEND_FS = 13

# Significance annotation font
SIG_FS = 12
SIG_LINEWIDTH = 1.2
SIG_NOTE_FS = 10

plt.rcParams.update({
    "font.size": TICK_FS,
    "axes.titlesize": TITLE_FS,
    "axes.labelsize": AXIS_LABEL_FS,
    "xtick.labelsize": TICK_FS,
    "ytick.labelsize": TICK_FS,
    "legend.fontsize": LEGEND_FS,
})

# ---------------------------------------------------------
# Statistical annotation display settings
# ---------------------------------------------------------
SHOW_NS_ROUND_PLOTS = False
SHOW_EXACT_P_ROUND_PLOTS = False

SIGNIFICANCE_NOTE = "ns: p ≥ 0.05, *: p < 0.05, **: p < 0.01, ***: p < 0.001, ****: p < 0.0001"

# ---------------------------------------------------------
# Round configurations
# ---------------------------------------------------------

ROUND_CONFIGS = {
    "round1": {
        "exposure": 0.005,
        "expected_total_runs": 5,
        "sources": {
            "normal": {
                "base_dir": Path("/Users/liuzimo/Desktop/MC data/Glas_NormalExpo/csv"),
                "files": [
                    "04Apr2026 normal test1-1.csv",
                    "04Apr2026 normal test4-2.csv",
                    "04Apr2026 normal test6-3.csv",
                    "04Apr2026 normal test8-4.csv",
                    "04Apr2026 normal test9-5.csv",
                ],
            }
        },
    },
    "round2": {
        "exposure": 0.01,
        "expected_total_runs": 5,
        "sources": {
            "normal": {
                "base_dir": Path("/Users/liuzimo/Desktop/MC data/Glas_DoubleExpo/csv"),
                "files": [
                    "04Apr2026 double test1-1.csv",
                    "04Apr2026 double test2-2.csv",
                    "04Apr2026 double test3-3.csv",
                    "04Apr2026 double test5-4.csv",
                    "04Apr2026 double test6-5.csv",
                ],
            }
        },
    },
    "round3": {
        "exposure": 0.015,
        "expected_total_runs": 5,
        "sources": {
            "normal": {
                "base_dir": Path("/Users/liuzimo/Desktop/MC data/Glas_TripleExpo/csv"),
                "files": [
                    "06Apr2026 triple test1-1.csv",
                    "06Apr2026 triple test2-2.csv",
                    "06Apr2026 triple test3-3.csv",
                    "06Apr2026 triple test4-4.csv",
                    "06Apr2026 triple test6-5.csv",
                ],
            }
        },
    },
    "round4": {
        "exposure": 0.02,
        "expected_total_runs": 5,
        "sources": {
            "normal": {
                "base_dir": Path("/Users/liuzimo/Desktop/MC data/Glas_QuadrupleExpo/csv"),
                "files": [
                    "08Apr2026 quadruple test1-1.csv",
                    "08Apr2026 quadruple test2-2.csv",
                    "08Apr2026 quadruple test3-3.csv",
                    "08Apr2026 quadruple test5-4.csv",
                    "08Apr2026 quadruple test6-5.csv",
                ],
            }
        },
    },
}

ROUND_ORDER = ["round1", "round2", "round3", "round4"]
CALIBRATION_ROUNDS = ["round1", "round2", "round3"]


# =========================================================
# MODEL SETTINGS
# =========================================================

# Based on the Glasgow example run you provided:
# Population=109, Susceptible=108, Exposed=1, Infected=0, Recovered=0
N = 109
E0, I0, R0 = 1, 0, 0
S0 = N - (E0 + I0 + R0)
y0 = (S0, E0, I0, R0)

PARAM_NAMES = ["beta", "epsilon", "gamma", "rho", "alpha"]
OBS_COLS = ["population", "susceptible", "exposed", "infected", "recovered", "dead"]

x0 = np.array([
    0.033,
    1 / 35,
    1 / 95,
    1 / 650,
    0.0,
], dtype=float)

lower_bounds = np.array([
    0.001,
    1 / 300,
    1 / 300,
    0.0,
    0.0,
], dtype=float)

upper_bounds = np.array([
    0.10,
    1 / 5,
    1 / 10,
    0.02,
    0.002,
], dtype=float)

BETA_LOWER = lower_bounds[0]
BETA_UPPER = upper_bounds[0]

COLORS = {
    "population": "black",
    "susceptible": "#B8E186",
    "exposed": "#F1B6DA",
    "infected": "#DE77AE",
    "recovered": "#41AB5D",
    "dead": "#636363",
}

LABELS = {
    "population": "Total Population",
    "susceptible": "Susceptible",
    "exposed": "Exposed",
    "infected": "Infectious",
    "recovered": "Recovered",
    "dead": "Dead",
}

warnings.filterwarnings("ignore", category=RuntimeWarning)
pd.set_option("display.max_columns", None)


# =========================================================
# HELPER FUNCTIONS
# =========================================================

def safe_round_label(round_name):
    return ROUND_LABELS.get(round_name, round_name)


def show_section(title):
    text = f"## {title}"
    if HAS_IPYTHON_DISPLAY:
        display(Markdown(text))
    else:
        print(f"\n{title}\n" + "-" * len(title))


def show_df(title, df, round_digits=4):
    show_section(title)
    if df is None or len(df) == 0:
        print("No data available.")
        return

    df_show = df.copy()
    num_cols = df_show.select_dtypes(include=[np.number]).columns
    df_show[num_cols] = df_show[num_cols].round(round_digits)

    if HAS_IPYTHON_DISPLAY:
        display(df_show)
    else:
        print(df_show.to_string(index=False))


def format_float_to_n_nonzero_decimal_digits(x, n_nonzero=2):
    if pd.isna(x):
        return ""

    if isinstance(x, (bool, np.bool_)):
        return str(x)

    if isinstance(x, (int, np.integer)) and not isinstance(x, (bool, np.bool_)):
        return str(int(x))

    try:
        d = Decimal(str(x))
    except (InvalidOperation, ValueError, TypeError):
        return str(x)

    if not d.is_finite():
        return str(x)

    if d == d.to_integral_value():
        return str(d.quantize(Decimal("1")))

    abs_d = abs(d)
    s = format(abs_d, "f")

    if "." not in s:
        return format(d, "f")

    _, frac_part = s.split(".")
    frac_trimmed = frac_part.rstrip("0")

    if frac_trimmed == "":
        return str(d.quantize(Decimal("1")))

    nonzero_positions = [i + 1 for i, ch in enumerate(frac_trimmed) if ch != "0"]

    if len(nonzero_positions) >= n_nonzero:
        places = nonzero_positions[n_nonzero - 1]
    else:
        places = len(frac_trimmed)

    quant = Decimal("1").scaleb(-places)
    rounded = d.quantize(quant, rounding=ROUND_HALF_UP)

    return format(rounded, f".{places}f")


def format_float_full_precision(x):
    if pd.isna(x):
        return ""

    if isinstance(x, (bool, np.bool_)):
        return str(x)

    if isinstance(x, (int, np.integer)) and not isinstance(x, (bool, np.bool_)):
        return str(int(x))

    try:
        return format(float(x), ".17g")
    except (ValueError, TypeError):
        return str(x)


def prepare_df_for_csv(df, n_nonzero=CSV_NONZERO_DECIMALS, preserve_precision=False):
    if df is None:
        return df

    df_out = df.copy()

    for col in df_out.columns:
        if pd.api.types.is_bool_dtype(df_out[col]):
            continue

        if pd.api.types.is_integer_dtype(df_out[col]):
            continue

        if pd.api.types.is_numeric_dtype(df_out[col]):
            if preserve_precision:
                df_out[col] = df_out[col].map(
                    lambda v: format_float_full_precision(v) if pd.notna(v) else ""
                )
            else:
                df_out[col] = df_out[col].map(
                    lambda v: format_float_to_n_nonzero_decimal_digits(v, n_nonzero=n_nonzero)
                    if pd.notna(v) else ""
                )

    return df_out


def save_csv(df, path, index=False, n_nonzero=CSV_NONZERO_DECIMALS, preserve_precision=False):
    df_out = prepare_df_for_csv(
        df,
        n_nonzero=n_nonzero,
        preserve_precision=preserve_precision
    )
    df_out.to_csv(path, index=index)


def finalize_figure(fig, outpath, dpi=300):
    fig.savefig(outpath, dpi=dpi, bbox_inches="tight")
    if SHOW_PLOT_PREVIEW:
        plt.show()
    plt.close(fig)


def rmse(a, b):
    a = np.asarray(a, dtype=float)
    b = np.asarray(b, dtype=float)
    mask = np.isfinite(a) & np.isfinite(b)
    if mask.sum() == 0:
        return np.nan
    return np.sqrt(np.mean((a[mask] - b[mask]) ** 2))


def mae(a, b):
    a = np.asarray(a, dtype=float)
    b = np.asarray(b, dtype=float)
    mask = np.isfinite(a) & np.isfinite(b)
    if mask.sum() == 0:
        return np.nan
    return np.mean(np.abs(a[mask] - b[mask]))


def corrcoef_safe(a, b):
    a = np.asarray(a, dtype=float)
    b = np.asarray(b, dtype=float)
    mask = np.isfinite(a) & np.isfinite(b)
    if mask.sum() < 2:
        return np.nan
    a2 = a[mask]
    b2 = b[mask]
    if np.allclose(np.std(a2), 0) or np.allclose(np.std(b2), 0):
        return np.nan
    return np.corrcoef(a2, b2)[0, 1]


def auc_trapz(t, y):
    t = np.asarray(t, dtype=float)
    y = np.asarray(y, dtype=float)
    mask = np.isfinite(t) & np.isfinite(y)
    if mask.sum() < 2:
        return np.nan
    t2 = t[mask]
    y2 = y[mask]
    order = np.argsort(t2)
    return np.trapz(y2[order], t2[order])


def interp_with_nan(x_new, x, y):
    x = np.asarray(x, dtype=float)
    y = np.asarray(y, dtype=float)
    x_new = np.asarray(x_new, dtype=float)

    if len(x) == 0:
        return np.full_like(x_new, np.nan, dtype=float)

    if len(x) == 1:
        out = np.full_like(x_new, np.nan, dtype=float)
        out[x_new == x[0]] = y[0]
        return out

    out = np.interp(x_new, x, y)
    out[(x_new < x.min()) | (x_new > x.max())] = np.nan
    return out


def nan_median_q1_q3(arr):
    arr = np.asarray(arr, dtype=float)
    med = np.nanmedian(arr, axis=0)
    q1 = np.nanpercentile(arr, 25, axis=0)
    q3 = np.nanpercentile(arr, 75, axis=0)
    return med, q1, q3


def mannwhitney_p(x, y):
    x = pd.Series(x).dropna().to_numpy()
    y = pd.Series(y).dropna().to_numpy()
    if len(x) == 0 or len(y) == 0:
        return np.nan
    return mannwhitneyu(x, y, alternative="two-sided").pvalue


def maybe_kruskal(groups):
    clean = [pd.Series(g).dropna().to_numpy() for g in groups]
    clean = [g for g in clean if len(g) > 0]
    if len(clean) < 2:
        return np.nan, np.nan
    stat, p = kruskal(*clean)
    return stat, p


def finite_values(arr):
    arr = np.asarray(arr, dtype=float)
    return arr[np.isfinite(arr)]


def finite_max(arr):
    vals = finite_values(arr)
    if len(vals) == 0:
        return np.nan
    return float(np.max(vals))


def flatten_finite(groups):
    vals = []
    for g in groups:
        gv = finite_values(g)
        if len(gv) > 0:
            vals.append(gv)
    if len(vals) == 0:
        return np.array([], dtype=float)
    return np.concatenate(vals)


def format_pvalue(p):
    if pd.isna(p):
        return "p = NA"
    if p < 1e-4:
        return "p < 1e-4"
    return f"p = {p:.4f}"


def p_to_sig_symbol(p, show_ns=True):
    if pd.isna(p):
        return "NA"
    if p < 1e-4:
        return "****"
    if p < 1e-3:
        return "***"
    if p < 1e-2:
        return "**"
    if p < 0.05:
        return "*"
    return "ns" if show_ns else ""


def make_sig_text(p, show_ns=True, show_exact_p=False):
    sig = p_to_sig_symbol(p, show_ns=show_ns)
    if sig == "":
        return ""
    if show_exact_p:
        return f"{sig}\n{format_pvalue(p)}"
    return sig


def add_sig_bracket(ax, x1, x2, y, h, text, linewidth=SIG_LINEWIDTH, fontsize=SIG_FS):
    if text is None or text == "":
        return
    ax.plot([x1, x1, x2, x2], [y, y + h, y + h, y], lw=linewidth, c="black", clip_on=False)
    ax.text((x1 + x2) / 2, y + h, text, ha="center", va="bottom", fontsize=fontsize)


def intervals_overlap(a1, a2, b1, b2):
    return not (a2 < b1 or b2 < a1)


def assign_annotation_levels(pair_records):
    if len(pair_records) == 0:
        return pair_records

    sorted_records = sorted(pair_records, key=lambda d: ((d["x2"] - d["x1"]), d["x1"], d["x2"]))
    placed = []

    for rec in sorted_records:
        level = 0
        while True:
            conflict = False
            for other in placed:
                if other["level"] == level and intervals_overlap(rec["x1"], rec["x2"], other["x1"], other["x2"]):
                    conflict = True
                    break
            if not conflict:
                break
            level += 1

        rec["level"] = level
        placed.append(rec)

    return placed


# =========================================================
# LOADING FUNCTIONS
# =========================================================

def load_run(path, round_name, source_name):
    df = pd.read_csv(path, encoding="utf-8-sig")
    df.columns = df.columns.str.strip().str.lower()

    required_cols = [
        "timestamp", "population", "susceptible",
        "exposed", "infected", "recovered", "dead"
    ]
    missing = [c for c in required_cols if c not in df.columns]
    if missing:
        raise ValueError(f"{path.name} missing columns: {missing}")

    df = df[required_cols].copy()

    ts = pd.to_datetime(df["timestamp"], format="%d%b%Y %H:%M:%S.%f", errors="coerce")
    if ts.isna().any():
        ts = pd.to_datetime(df["timestamp"], errors="coerce")

    df["timestamp"] = ts
    df = df.dropna(subset=["timestamp"]).copy()

    df["t"] = (df["timestamp"] - df["timestamp"].iloc[0]).dt.total_seconds()

    df[OBS_COLS] = df[OBS_COLS].apply(pd.to_numeric, errors="coerce")
    df = df.dropna(subset=OBS_COLS).copy()

    df = df.groupby("t", as_index=False)[OBS_COLS].mean()
    df = df[(df["t"] >= 0) & (df["t"] <= RUN_END)].sort_values("t").reset_index(drop=True)

    if len(df) < 2:
        raise ValueError(f"{path.name} has fewer than 2 usable rows.")

    df["round"] = round_name
    df["source"] = source_name
    df["run"] = path.stem

    return df[["round", "source", "run", "t"] + OBS_COLS]


def load_round_dataset(round_name, config):
    normal_runs = []
    missing_files = []

    for source_name, source_cfg in config["sources"].items():
        base_dir = Path(source_cfg["base_dir"])
        files = source_cfg.get("files", [])

        if len(files) == 0:
            continue

        for fname in files:
            path = base_dir / fname
            if not path.exists():
                missing_files.append(str(path))
                continue

            run_df = load_run(path, round_name, source_name)

            if source_name == "normal":
                normal_runs.append(run_df)
            else:
                raise ValueError(f"Unknown source name: {source_name}")

    pooled_runs = normal_runs

    return {
        "round": round_name,
        "exposure": config["exposure"],
        "expected_total_runs": config["expected_total_runs"],
        "normal_runs": normal_runs,
        "pooled_runs": pooled_runs,
        "missing_files": missing_files,
    }


def make_coverage_table(runs):
    rows = []
    expected_points = len(COMMON_GRID)

    for run_df in runs:
        rows.append({
            "round": run_df["round"].iloc[0],
            "source": run_df["source"].iloc[0],
            "run": run_df["run"].iloc[0],
            "points_recorded": len(run_df),
            "last_recorded_second": run_df["t"].max(),
            "coverage_fraction": len(run_df) / expected_points,
        })

    return pd.DataFrame(rows)


# =========================================================
# OBSERVED DATA SUMMARIES
# =========================================================

def build_observed_stats(runs):
    if len(runs) == 0:
        raise ValueError("No runs provided to build_observed_stats().")

    obs_stacks = {c: [] for c in OBS_COLS}

    for run_df in runs:
        x = run_df["t"].to_numpy()
        for c in OBS_COLS:
            obs_stacks[c].append(interp_with_nan(COMMON_GRID, x, run_df[c].to_numpy()))

    stats = {}
    for c in OBS_COLS:
        arr = np.vstack(obs_stacks[c])
        med, q1, q3 = nan_median_q1_q3(arr)
        stats[c] = {
            "stack": arr,
            "median": med,
            "q1": q1,
            "q3": q3,
        }

    return stats


def observed_stats_to_df(round_name, obs_stats):
    rows = []
    for i, t in enumerate(COMMON_GRID):
        row = {"round": round_name, "t": t}
        for c in OBS_COLS:
            row[f"{c}_median"] = obs_stats[c]["median"][i]
            row[f"{c}_q1"] = obs_stats[c]["q1"][i]
            row[f"{c}_q3"] = obs_stats[c]["q3"][i]
        rows.append(row)
    return pd.DataFrame(rows)


# =========================================================
# AUC / EPIDEMIC SUMMARY FUNCTIONS
# =========================================================

def build_per_run_auc_table(runs):
    rows = []

    for run_df in runs:
        t = run_df["t"].to_numpy()
        E = run_df["exposed"].to_numpy()
        I = run_df["infected"].to_numpy()

        rows.append({
            "round": run_df["round"].iloc[0],
            "source": run_df["source"].iloc[0],
            "run": run_df["run"].iloc[0],
            "n_points": len(run_df),
            "last_second": float(np.nanmax(t)),
            "auc_infected": auc_trapz(t, I),
            "auc_exposed_infected": auc_trapz(t, E + I),
            "peak_infected": float(np.nanmax(I)),
            "time_to_peak_infected": float(t[np.nanargmax(I)]) if len(I) > 0 else np.nan,
        })

    return pd.DataFrame(rows)


def build_round_comparison_table(auc_df):
    rows = []
    metrics = ["auc_infected", "auc_exposed_infected", "peak_infected"]
    rounds_present = [r for r in ROUND_ORDER if r in auc_df["round"].unique()]

    for metric in metrics:
        grouped = [
            auc_df.loc[auc_df["round"] == r, metric].dropna().to_numpy()
            for r in rounds_present
        ]

        stat, p = maybe_kruskal(grouped)
        rows.append({
            "comparison_type": "overall_kruskal",
            "metric": metric,
            "group1": ",".join(rounds_present),
            "group2": "",
            "statistic": stat,
            "p_value": p,
        })

        for r1, r2 in combinations(rounds_present, 2):
            x = auc_df.loc[auc_df["round"] == r1, metric].dropna().to_numpy()
            y = auc_df.loc[auc_df["round"] == r2, metric].dropna().to_numpy()

            rows.append({
                "comparison_type": "pairwise_mannwhitney",
                "metric": metric,
                "group1": r1,
                "group2": r2,
                "statistic": np.nan,
                "p_value": mannwhitney_p(x, y),
            })

    return pd.DataFrame(rows)


def build_round_auc_summary(auc_df):
    return (
        auc_df.groupby("round", as_index=False)
        .agg(
            n_runs=("run", "count"),
            exposure=("round", lambda s: ROUND_CONFIGS[s.iloc[0]]["exposure"]),
            mean_auc_infected=("auc_infected", "mean"),
            median_auc_infected=("auc_infected", "median"),
            mean_auc_exposed_infected=("auc_exposed_infected", "mean"),
            median_auc_exposed_infected=("auc_exposed_infected", "median"),
            mean_peak_infected=("peak_infected", "mean"),
            median_peak_infected=("peak_infected", "median"),
        )
        .sort_values("round")
        .reset_index(drop=True)
    )


def build_ode_vs_mc_auc_distribution_summary(round_data, fitted_params_by_round):
    rows = []

    for round_name in ROUND_ORDER:
        if round_name not in round_data:
            continue
        if round_name not in fitted_params_by_round:
            continue
        if len(round_data[round_name].get("pooled_runs", [])) == 0:
            continue

        runs = round_data[round_name]["pooled_runs"]
        params = fitted_params_by_round[round_name]

        mc_auc_vals = []
        for run_df in runs:
            auc_val = auc_trapz(run_df["t"].to_numpy(), run_df["infected"].to_numpy())
            if pd.notna(auc_val):
                mc_auc_vals.append(float(auc_val))

        if len(mc_auc_vals) == 0:
            continue

        mc_auc_vals = np.asarray(mc_auc_vals, dtype=float)

        fit_curve = simulate_curve_dict(params, COMMON_GRID)
        ode_auc = auc_trapz(COMMON_GRID, fit_curve["infected"])

        mc_min = float(np.min(mc_auc_vals))
        mc_q1 = float(np.percentile(mc_auc_vals, 25))
        mc_median = float(np.median(mc_auc_vals))
        mc_q3 = float(np.percentile(mc_auc_vals, 75))
        mc_max = float(np.max(mc_auc_vals))

        rows.append({
            "round": round_name,
            "exposure": ROUND_CONFIGS[round_name]["exposure"],
            "n_mc_runs": len(mc_auc_vals),
            "mc_auc_min": mc_min,
            "mc_auc_q1": mc_q1,
            "mc_auc_median": mc_median,
            "mc_auc_q3": mc_q3,
            "mc_auc_max": mc_max,
            "ode_auc_infected": float(ode_auc),
            "ode_minus_mc_median": float(ode_auc - mc_median),
            "ode_abs_diff_from_mc_median": float(abs(ode_auc - mc_median)),
            "ode_within_mc_iqr": bool(mc_q1 <= ode_auc <= mc_q3),
            "ode_within_mc_range": bool(mc_min <= ode_auc <= mc_max),
        })

    return pd.DataFrame(rows)


# =========================================================
# ODE MODEL
# =========================================================

def deriv(y, t, beta, epsilon, gamma, rho, alpha):
    S, E, I, R = y
    Nt = max(S + E + I + R, 1e-9)

    dSdt = (-beta * S * I / Nt) + (alpha * R)
    dEdt = (beta * S * I / Nt) - (epsilon * E)
    dIdt = epsilon * E - ((gamma + rho) * I)
    dRdt = gamma * I - (alpha * R)

    return dSdt, dEdt, dIdt, dRdt


def simulate(params, tgrid):
    beta, epsilon, gamma, rho, alpha = params

    sol = odeint(
        deriv,
        y0,
        tgrid,
        args=(beta, epsilon, gamma, rho, alpha),
        mxstep=10000
    )

    S, E, I, R = sol.T
    P = S + E + I + R
    D = np.clip(N - P, 0, None)

    return S, E, I, R, P, D


def simulate_curve_dict(params, tgrid):
    S, E, I, R, P, D = simulate(params, tgrid)
    return {
        "population": P,
        "susceptible": S,
        "exposed": E,
        "infected": I,
        "recovered": R,
        "dead": D,
    }


# =========================================================
# RESIDUALS AND FITTING
# =========================================================

def residuals_one_run(params, run_df):
    t_obs = run_df["t"].to_numpy()

    try:
        S, E, I, R, P, D = simulate(params, t_obs)
    except Exception:
        return np.full(5 * len(t_obs), 1e6)

    if not (np.all(np.isfinite(S)) and np.all(np.isfinite(E)) and np.all(np.isfinite(I)) and np.all(np.isfinite(R))):
        return np.full(5 * len(t_obs), 1e6)

    if np.any(S < -1e-6) or np.any(E < -1e-6) or np.any(I < -1e-6) or np.any(R < -1e-6):
        return np.full(5 * len(t_obs), 1e6)

    S_obs = run_df["susceptible"].to_numpy()
    E_obs = run_df["exposed"].to_numpy()
    I_obs = run_df["infected"].to_numpy()
    R_obs = run_df["recovered"].to_numpy()
    P_obs = run_df["population"].to_numpy()

    scale_S = max(np.nanmax(S_obs), 1.0)
    scale_E = max(np.nanmax(E_obs), 1.0)
    scale_I = max(np.nanmax(I_obs), 1.0)
    scale_R = max(np.nanmax(R_obs), 1.0)
    scale_P = max(np.nanmax(P_obs), 1.0)

    res_S = 0.8 * (S - S_obs) / scale_S
    res_E = 1.2 * (E - E_obs) / scale_E
    res_I = 1.2 * (I - I_obs) / scale_I
    res_R = 1.0 * (R - R_obs) / scale_R
    res_P = 0.8 * (P - P_obs) / scale_P

    res = np.concatenate([res_S, res_E, res_I, res_R, res_P])
    res = res / np.sqrt(len(t_obs))

    return res


def pooled_residuals(params, runs):
    total_len = sum(5 * len(run_df) for run_df in runs)
    all_res = []

    for run_df in runs:
        res = residuals_one_run(params, run_df)
        if not np.all(np.isfinite(res)):
            return np.full(total_len, 1e6)
        all_res.append(res)

    return np.concatenate(all_res)


def pooled_residuals_beta_only(beta_arr, base_params, runs):
    params = np.array(base_params, dtype=float).copy()
    params[0] = float(beta_arr[0])
    return pooled_residuals(params, runs)


def fit_full_params(round_name, runs):
    fit = least_squares(
        lambda p: pooled_residuals(p, runs),
        x0=x0,
        bounds=(lower_bounds, upper_bounds),
        method="trf",
        loss="soft_l1",
        f_scale=0.2,
        max_nfev=1500,
    )

    best_params = fit.x.copy()

    fit_row = {
        "round": round_name,
        "fit_mode": "full_5_parameter_fit",
        "success": bool(fit.success),
        "message": fit.message,
        "cost": float(fit.cost),
    }
    fit_row.update({name: value for name, value in zip(PARAM_NAMES, best_params)})

    return best_params, fit, pd.DataFrame([fit_row])


def fit_beta_only(round_name, runs, base_params, beta_init):
    lower = np.array([BETA_LOWER], dtype=float)
    upper = np.array([BETA_UPPER], dtype=float)
    x0_beta = np.array([float(beta_init)], dtype=float)

    fit = least_squares(
        lambda b: pooled_residuals_beta_only(b, base_params, runs),
        x0=x0_beta,
        bounds=(lower, upper),
        method="trf",
        loss="soft_l1",
        f_scale=0.2,
        max_nfev=1200,
    )

    params = np.array(base_params, dtype=float).copy()
    params[0] = fit.x[0]

    fit_row = {
        "round": round_name,
        "fit_mode": "beta_only_fit_from_round1_baseline",
        "success": bool(fit.success),
        "message": fit.message,
        "cost": float(fit.cost),
        "beta": float(params[0]),
        "epsilon_fixed": float(params[1]),
        "gamma_fixed": float(params[2]),
        "rho_fixed": float(params[3]),
        "alpha_fixed": float(params[4]),
    }

    return params, fit, pd.DataFrame([fit_row])


# =========================================================
# DIAGNOSTICS / VALIDATION
# =========================================================

def per_run_diagnostics(params, runs, fit_mode):
    rows = []

    for run_df in runs:
        t_obs = run_df["t"].to_numpy()
        S_fit, E_fit, I_fit, R_fit, P_fit, D_fit = simulate(params, t_obs)

        E_obs = run_df["exposed"].to_numpy()
        I_obs = run_df["infected"].to_numpy()

        rows.append({
            "round": run_df["round"].iloc[0],
            "source": run_df["source"].iloc[0],
            "run": run_df["run"].iloc[0],
            "fit_mode": fit_mode,
            "n_points": len(run_df),
            "rmse_population": rmse(P_fit, run_df["population"].to_numpy()),
            "rmse_susceptible": rmse(S_fit, run_df["susceptible"].to_numpy()),
            "rmse_exposed": rmse(E_fit, E_obs),
            "rmse_infected": rmse(I_fit, I_obs),
            "rmse_recovered": rmse(R_fit, run_df["recovered"].to_numpy()),
            "rmse_dead_external_check": rmse(D_fit, run_df["dead"].to_numpy()),
            "auc_obs_infected": auc_trapz(t_obs, I_obs),
            "auc_fit_infected": auc_trapz(t_obs, I_fit),
            "auc_obs_exposed_infected": auc_trapz(t_obs, E_obs + I_obs),
            "auc_fit_exposed_infected": auc_trapz(t_obs, E_fit + I_fit),
            "peak_obs_infected": float(np.nanmax(I_obs)),
            "peak_fit_infected": float(np.nanmax(I_fit)),
        })

    return pd.DataFrame(rows)


def pooled_curve_validation(round_name, exposure, obs_stats, fit_curve, fit_mode):
    E_obs = obs_stats["exposed"]["median"]
    I_obs = obs_stats["infected"]["median"]
    EI_obs = E_obs + I_obs

    E_fit = fit_curve["exposed"]
    I_fit = fit_curve["infected"]
    EI_fit = E_fit + I_fit

    row = {
        "round": round_name,
        "exposure": exposure,
        "fit_mode": fit_mode,
        "auc_obs_median_infected": auc_trapz(COMMON_GRID, I_obs),
        "auc_model_infected": auc_trapz(COMMON_GRID, I_fit),
        "auc_abs_diff_infected": np.abs(auc_trapz(COMMON_GRID, I_fit) - auc_trapz(COMMON_GRID, I_obs)),
        "auc_obs_median_exposed_infected": auc_trapz(COMMON_GRID, EI_obs),
        "auc_model_exposed_infected": auc_trapz(COMMON_GRID, EI_fit),
        "auc_abs_diff_exposed_infected": np.abs(auc_trapz(COMMON_GRID, EI_fit) - auc_trapz(COMMON_GRID, EI_obs)),
        "rmse_median_population": rmse(fit_curve["population"], obs_stats["population"]["median"]),
        "rmse_median_susceptible": rmse(fit_curve["susceptible"], obs_stats["susceptible"]["median"]),
        "rmse_median_exposed": rmse(fit_curve["exposed"], obs_stats["exposed"]["median"]),
        "rmse_median_infected": rmse(fit_curve["infected"], obs_stats["infected"]["median"]),
        "rmse_median_recovered": rmse(fit_curve["recovered"], obs_stats["recovered"]["median"]),
        "peak_obs_infected_median": np.nanmax(I_obs),
        "peak_model_infected": np.nanmax(I_fit),
        "time_to_peak_obs_infected_median": COMMON_GRID[np.nanargmax(I_obs)],
        "time_to_peak_model_infected": COMMON_GRID[np.nanargmax(I_fit)],
    }

    return pd.DataFrame([row])


def fit_closeness_summary(round_name, exposure, obs_stats, fit_curve, fit_mode):
    rows = []
    compartments = ["population", "susceptible", "exposed", "infected", "recovered"]

    for c in compartments:
        obs = obs_stats[c]["median"]
        fit = fit_curve[c]

        row = {
            "round": round_name,
            "exposure": exposure,
            "fit_mode": fit_mode,
            "compartment": c,
            "rmse": rmse(fit, obs),
            "mae": mae(fit, obs),
            "correlation": corrcoef_safe(fit, obs),
            "auc_obs": auc_trapz(COMMON_GRID, obs),
            "auc_model": auc_trapz(COMMON_GRID, fit),
            "auc_abs_diff": abs(auc_trapz(COMMON_GRID, fit) - auc_trapz(COMMON_GRID, obs)),
        }

        if c == "infected":
            row["peak_obs"] = np.nanmax(obs)
            row["peak_model"] = np.nanmax(fit)
            row["peak_abs_diff"] = abs(np.nanmax(fit) - np.nanmax(obs))
            row["time_to_peak_obs"] = COMMON_GRID[np.nanargmax(obs)]
            row["time_to_peak_model"] = COMMON_GRID[np.nanargmax(fit)]
            row["time_to_peak_abs_diff"] = abs(
                COMMON_GRID[np.nanargmax(fit)] - COMMON_GRID[np.nanargmax(obs)]
            )
        else:
            row["peak_obs"] = np.nan
            row["peak_model"] = np.nan
            row["peak_abs_diff"] = np.nan
            row["time_to_peak_obs"] = np.nan
            row["time_to_peak_model"] = np.nan
            row["time_to_peak_abs_diff"] = np.nan

        rows.append(row)

    return pd.DataFrame(rows)


# =========================================================
# PLOTTING
# =========================================================

def plot_round_auc_boxplots(auc_df, outpath):
    if auc_df.empty:
        return

    fig, axes = plt.subplots(1, 2, figsize=(15, 7), facecolor="white")
    metrics = [
        ("auc_infected", f"{CITY_NAME} comparison across rounds: AUC of Infected"),
        ("auc_exposed_infected", f"{CITY_NAME} comparison across rounds: AUC of Exposed + Infected"),
    ]

    round_names_present = [r for r in ROUND_ORDER if r in auc_df["round"].unique()]

    for ax, (metric, title) in zip(axes, metrics):
        groups = []
        labels = []
        group_records = []

        for round_name in round_names_present:
            vals = auc_df.loc[auc_df["round"] == round_name, metric].dropna().to_numpy()
            if len(vals) > 0:
                groups.append(vals)
                labels.append(safe_round_label(round_name))
                group_records.append({
                    "round": round_name,
                    "values": vals,
                    "position": len(groups),
                })

        if len(groups) == 0:
            continue

        bp = ax.boxplot(groups, patch_artist=True, labels=labels, showmeans=True)
        colors = ["#c6dbef", "#9ecae1", "#6baed6", "#3182bd"]
        for patch, color in zip(bp["boxes"], colors[:len(groups)]):
            patch.set_facecolor(color)

        ax.set_title(title, fontsize=BOXPLOT_TITLE_FS)
        ax.set_ylabel("AUC", fontsize=AXIS_LABEL_FS)
        ax.tick_params(axis="x", labelsize=BOXPLOT_XTICK_FS)
        ax.tick_params(axis="y", labelsize=BOXPLOT_YTICK_FS)

        stat, overall_p = maybe_kruskal([g["values"] for g in group_records])
        overall_sig = p_to_sig_symbol(overall_p, show_ns=True)
        overall_text = (
            f"Overall Kruskal–Wallis: {overall_sig}, {format_pvalue(overall_p)}\n"
            f"Pairwise: Mann–Whitney U"
        )

        ax.text(
            0.98, 0.02,
            overall_text,
            transform=ax.transAxes,
            ha="right", va="bottom",
            fontsize=SIG_NOTE_FS,
            bbox=dict(boxstyle="round,pad=0.25", facecolor="white", edgecolor="none", alpha=0.85)
        )

        all_vals = flatten_finite(groups)
        if len(all_vals) == 0:
            continue

        data_max = np.max(all_vals)
        data_min = np.min(all_vals)
        y_range = max(data_max - data_min, 1.0)

        pair_records = []
        for rec1, rec2 in combinations(group_records, 2):
            p = mannwhitney_p(rec1["values"], rec2["values"])
            sig_text = make_sig_text(
                p,
                show_ns=SHOW_NS_ROUND_PLOTS,
                show_exact_p=SHOW_EXACT_P_ROUND_PLOTS
            )

            if sig_text == "":
                continue

            pair_records.append({
                "x1": rec1["position"],
                "x2": rec2["position"],
                "p": p,
                "text": sig_text,
            })

        pair_records = assign_annotation_levels(pair_records)

        base_y = data_max + 0.06 * y_range
        step_y = 0.09 * y_range
        h = 0.02 * y_range

        y_upper_needed = data_max

        for rec in pair_records:
            y = base_y + rec["level"] * step_y
            add_sig_bracket(
                ax,
                rec["x1"],
                rec["x2"],
                y,
                h,
                rec["text"],
                linewidth=SIG_LINEWIDTH,
                fontsize=SIG_FS
            )
            y_upper_needed = max(y_upper_needed, y + h + 0.06 * y_range)

        current_bottom, current_top = ax.get_ylim()
        ax.set_ylim(current_bottom, max(current_top, y_upper_needed))

    fig.text(0.5, 0.02, SIGNIFICANCE_NOTE, ha="center", fontsize=SIG_NOTE_FS)
    plt.tight_layout(rect=[0, 0.06, 1, 0.98])
    finalize_figure(fig, outpath)


def plot_round_mc_only(round_name, exposure, obs_stats, outpath):
    fig, ax = plt.subplots(figsize=(12, 7), facecolor="white")
    ax.set_facecolor("white")
    ax.set_axisbelow(True)

    order = ["population", "susceptible", "exposed", "infected", "recovered", "dead"]

    for c in order:
        ax.fill_between(
            COMMON_GRID,
            obs_stats[c]["q1"],
            obs_stats[c]["q3"],
            color=COLORS[c],
            alpha=0.15
        )
        ax.plot(
            COMMON_GRID,
            obs_stats[c]["median"],
            color=COLORS[c],
            lw=3 if c != "population" else 2.5,
            alpha=0.9
        )

    ymax = np.nanmax(obs_stats["population"]["q3"]) * 1.15
    ax.set_xlim(0, RUN_END)
    ax.set_ylim(0, ymax)
    ax.set_xlabel("Time (seconds)", fontsize=AXIS_LABEL_FS)
    ax.set_ylabel("Number of Individuals", fontsize=AXIS_LABEL_FS)
    ax.set_title(
        f"{CITY_NAME} {safe_round_label(round_name)} (exposure = {exposure}): Minecraft observed data\nMedian + IQR",
        fontsize=TITLE_FS,
        pad=12
    )
    ax.tick_params(axis="both", labelsize=TICK_FS)

    legend_handles = [
        Line2D([0], [0], color=COLORS["population"], lw=2.5, label="Population"),
        Line2D([0], [0], color=COLORS["susceptible"], lw=3, label="Susceptible"),
        Line2D([0], [0], color=COLORS["exposed"], lw=3, label="Exposed"),
        Line2D([0], [0], color=COLORS["infected"], lw=3, label="Infectious"),
        Line2D([0], [0], color=COLORS["recovered"], lw=3, label="Recovered"),
        Line2D([0], [0], color=COLORS["dead"], lw=3, label="Dead"),
    ]
    ax.legend(
        handles=legend_handles,
        frameon=False,
        loc="upper left",
        bbox_to_anchor=(1.02, 1.0),
        borderaxespad=0,
        fontsize=OUTSIDE_LEGEND_FS
    )
    ax.grid(False)

    plt.tight_layout(rect=[0, 0, 0.82, 1])
    finalize_figure(fig, outpath)


def plot_round_ode_only(round_name, exposure, fit_curve, plot_label, outpath):
    fig, ax = plt.subplots(figsize=(12, 7), facecolor="white")
    ax.set_facecolor("white")
    ax.set_axisbelow(True)

    order = ["population", "susceptible", "exposed", "infected", "recovered", "dead"]

    for c in order:
        ax.plot(
            COMMON_GRID,
            fit_curve[c],
            color=COLORS[c],
            lw=3 if c != "population" else 2.5,
            alpha=0.95
        )

    ymax = np.nanmax(fit_curve["population"]) * 1.15
    ax.set_xlim(0, RUN_END)
    ax.set_ylim(0, ymax)
    ax.set_xlabel("Time (seconds)", fontsize=AXIS_LABEL_FS)
    ax.set_ylabel("Number of Individuals", fontsize=AXIS_LABEL_FS)
    ax.set_title(
        f"{CITY_NAME} {safe_round_label(round_name)} (exposure = {exposure}): ODE trajectories\n{plot_label}",
        fontsize=TITLE_FS,
        pad=12
    )
    ax.tick_params(axis="both", labelsize=TICK_FS)

    legend_handles = [
        Line2D([0], [0], color=COLORS["population"], lw=2.5, label="Population"),
        Line2D([0], [0], color=COLORS["susceptible"], lw=3, label="Susceptible"),
        Line2D([0], [0], color=COLORS["exposed"], lw=3, label="Exposed"),
        Line2D([0], [0], color=COLORS["infected"], lw=3, label="Infectious"),
        Line2D([0], [0], color=COLORS["recovered"], lw=3, label="Recovered"),
        Line2D([0], [0], color=COLORS["dead"], lw=3, label="Dead"),
    ]
    ax.legend(
        handles=legend_handles,
        frameon=False,
        loc="upper left",
        bbox_to_anchor=(1.02, 1.0),
        borderaxespad=0,
        fontsize=OUTSIDE_LEGEND_FS
    )
    ax.grid(False)

    plt.tight_layout(rect=[0, 0, 0.82, 1])
    finalize_figure(fig, outpath)


def plot_beta_exposure_trend(exposures, betas, beta4_pred, outpath, beta4_fit=None):
    fig, ax = plt.subplots(figsize=(14, 14), facecolor="white")

    exposures = np.asarray(exposures, dtype=float)
    betas = np.asarray(betas, dtype=float)

    coef = np.polyfit(exposures, betas, 1)
    xline = np.linspace(exposures.min(), 0.02, 200)
    yline = np.polyval(coef, xline)

    ax.plot(xline, yline, color="#2c7fb8", lw=2.5, label="Linear fit from Rounds 1-3")
    ax.scatter(exposures, betas, s=180, color="#d95f0e", zorder=3, label="Fitted beta (Rounds 1-3)")
    ax.scatter([0.02], [beta4_pred], s=240, color="#31a354", marker="*", zorder=4, label="Predicted beta for Round 4")

    y_all = list(betas) + [beta4_pred] + list(yline)

    if beta4_fit is not None and np.isfinite(beta4_fit):
        ax.scatter([0.02], [beta4_fit], s=180, color="#756bb1", marker="D", zorder=4, label="Fitted beta for Round 4")
        y_all.append(beta4_fit)

    x_min = min(np.min(exposures), 0.02, np.min(xline))
    x_max = max(np.max(exposures), 0.02, np.max(xline))
    y_min = np.min(y_all)
    y_max = np.max(y_all)

    x_span = x_max - x_min
    y_span = y_max - y_min
    base_span = max(x_span, y_span)

    pad = 0.08 * base_span
    half_span = 0.5 * base_span + pad

    x_center = 0.5 * (x_min + x_max)
    y_center = 0.5 * (y_min + y_max)

    ax.set_xlim(x_center - half_span, x_center + half_span)
    ax.set_ylim(y_center - half_span, y_center + half_span)

    ax.set_box_aspect(1)

    ax.set_xlabel("Minecraft exposure setting", fontsize=AXIS_LABEL_FS)
    ax.set_ylabel("ODE beta", fontsize=AXIS_LABEL_FS)
    ax.set_title(f"{CITY_NAME}: beta vs Minecraft exposure setting", fontsize=TITLE_FS)
    ax.tick_params(axis="both", labelsize=TICK_FS)
    ax.grid(False)

    ax.legend(
        frameon=False,
        fontsize=OUTSIDE_LEGEND_FS,
        loc="upper left",
        bbox_to_anchor=(1.02, 1.0),
        borderaxespad=0
    )

    plt.tight_layout(rect=[0, 0, 0.8, 1])
    finalize_figure(fig, outpath)


def plot_ode_vs_mc_auc_distribution(round_data, fitted_params_by_round, outpath):
    """
    Publication-style boxplot:
    - boxplots = MC replicate infected AUC per round
    - red diamond = fitted ODE infected AUC per round
    """
    groups = []
    labels = []
    ode_points_x = []
    ode_points_y = []

    for round_name in ROUND_ORDER:
        if round_name not in round_data:
            continue
        if round_name not in fitted_params_by_round:
            continue
        if len(round_data[round_name].get("pooled_runs", [])) == 0:
            continue

        runs = round_data[round_name]["pooled_runs"]
        params = fitted_params_by_round[round_name]

        mc_auc_vals = []
        for run_df in runs:
            auc_val = auc_trapz(run_df["t"].to_numpy(), run_df["infected"].to_numpy())
            if pd.notna(auc_val):
                mc_auc_vals.append(float(auc_val))

        if len(mc_auc_vals) == 0:
            continue

        groups.append(np.asarray(mc_auc_vals, dtype=float))
        labels.append(safe_round_label(round_name))

        fit_curve = simulate_curve_dict(params, COMMON_GRID)
        ode_auc = auc_trapz(COMMON_GRID, fit_curve["infected"])

        ode_points_x.append(len(groups))
        ode_points_y.append(float(ode_auc))

    if len(groups) == 0:
        return

    fig, ax = plt.subplots(figsize=(10, 7), facecolor="white")

    bp = ax.boxplot(
        groups,
        patch_artist=True,
        labels=labels,
        showmeans=True
    )

    box_colors = ["#c6dbef", "#9ecae1", "#6baed6", "#3182bd"]
    for patch, color in zip(bp["boxes"], box_colors[:len(groups)]):
        patch.set_facecolor(color)

    ax.scatter(
        ode_points_x,
        ode_points_y,
        s=140,
        color="#cb181d",
        marker="D",
        zorder=5,
        label="Fitted ODE infected AUC"
    )

    ax.set_xlabel("Round", fontsize=AXIS_LABEL_FS)
    ax.set_ylabel("AUC of infected", fontsize=AXIS_LABEL_FS)
    ax.set_title(f"{CITY_NAME}: ODE vs Minecraft infected AUC distribution by round", fontsize=TITLE_FS)
    ax.tick_params(axis="x", labelsize=TICK_FS)
    ax.tick_params(axis="y", labelsize=TICK_FS)
    ax.grid(False)

    legend_handles = [
        Line2D([0], [0], marker="D", color="w",
               markerfacecolor="#cb181d", markersize=10,
               label="Fitted ODE infected AUC")
    ]
    ax.legend(handles=legend_handles, frameon=False, loc="lower right", fontsize=LEGEND_FS)

    plt.tight_layout()
    finalize_figure(fig, outpath)


# =========================================================
# MAIN WORKFLOW
# =========================================================

def main():
    print(f"Loading {CITY_NAME} Minecraft data...")

    round_data = {}
    coverage_tables = []
    per_run_auc_tables = []
    observed_stat_tables = []

    for round_name in ROUND_ORDER:
        cfg = ROUND_CONFIGS[round_name]
        rd = load_round_dataset(round_name, cfg)

        if rd["missing_files"]:
            show_section(f"Missing files in {round_name}")
            for m in rd["missing_files"]:
                print(m)

        if len(rd["pooled_runs"]) == 0:
            print(f"{round_name}: no usable data loaded. Skipped.")
            round_data[round_name] = rd
            continue

        round_data[round_name] = rd

        coverage_df = make_coverage_table(rd["pooled_runs"])
        coverage_tables.append(coverage_df)

        auc_df = build_per_run_auc_table(rd["pooled_runs"])
        per_run_auc_tables.append(auc_df)

        obs_stats = build_observed_stats(rd["pooled_runs"])
        rd["obs_stats"] = obs_stats
        observed_stat_tables.append(observed_stats_to_df(round_name, obs_stats))

    if coverage_tables:
        coverage_all = pd.concat(coverage_tables, ignore_index=True)
        save_csv(coverage_all, CSV_OUT / "01_coverage_summary_all_rounds.csv", index=False)
    else:
        coverage_all = pd.DataFrame()

    if observed_stat_tables:
        obs_stats_all_df = pd.concat(observed_stat_tables, ignore_index=True)
        save_csv(obs_stats_all_df, CSV_OUT / "02_observed_median_q1_q3_all_rounds.csv", index=False)

    if per_run_auc_tables:
        auc_all = pd.concat(per_run_auc_tables, ignore_index=True)
        save_csv(auc_all, CSV_OUT / "03_per_run_auc_all_rounds.csv", index=False)
    else:
        auc_all = pd.DataFrame()

    # Coverage
    show_df("Coverage summary for all loaded runs", coverage_all)

    # Round-to-round MC AUC comparison
    pooled_auc_df = auc_all.copy()
    round_auc_summary_df = build_round_auc_summary(pooled_auc_df)
    round_auc_tests_df = build_round_comparison_table(pooled_auc_df)

    save_csv(round_auc_summary_df, CSV_OUT / "05_round_auc_summary.csv", index=False)
    save_csv(round_auc_tests_df, CSV_OUT / "06_round_auc_tests.csv", index=False)

    show_df(f"{CITY_NAME} round-to-round Minecraft AUC summary", round_auc_summary_df)
    show_df(f"{CITY_NAME} round-to-round Minecraft statistical tests", round_auc_tests_df)
    plot_round_auc_boxplots(
        pooled_auc_df,
        PLOT_OUT / "02_round_comparison_auc_boxplots.png"
    )

    # Round 1 full fit
    if len(round_data["round1"]["pooled_runs"]) == 0:
        raise RuntimeError("Round 1 data are required for the baseline full fit.")

    round1_params, round1_fit_obj, round1_fit_df = fit_full_params(
        "round1",
        round_data["round1"]["pooled_runs"]
    )
    save_csv(round1_fit_df, CSV_OUT / "07_round1_full_fit_parameters.csv", index=False)

    fitted_params_by_round = {"round1": round1_params.copy()}
    fit_tables = [round1_fit_df]
    diag_tables = []
    curve_validation_tables = []
    closeness_tables = []

    round1_diag = per_run_diagnostics(
        round1_params,
        round_data["round1"]["pooled_runs"],
        fit_mode="full_5_parameter_fit"
    )
    diag_tables.append(round1_diag)

    round1_curve = simulate_curve_dict(round1_params, COMMON_GRID)
    save_csv(pd.DataFrame({"t": COMMON_GRID, **round1_curve}), CSV_OUT / "08_round1_fitted_curve.csv", index=False)

    round1_val = pooled_curve_validation(
        "round1",
        ROUND_CONFIGS["round1"]["exposure"],
        round_data["round1"]["obs_stats"],
        round1_curve,
        fit_mode="full_5_parameter_fit"
    )
    curve_validation_tables.append(round1_val)

    round1_close_df = fit_closeness_summary(
        "round1",
        ROUND_CONFIGS["round1"]["exposure"],
        round_data["round1"]["obs_stats"],
        round1_curve,
        "full_5_parameter_fit"
    )
    closeness_tables.append(round1_close_df)
    save_csv(round1_close_df, CSV_OUT / "09_round1_fit_closeness_summary.csv", index=False)

    show_df("Round 1 full 5-parameter fit", round1_fit_df)
    show_df("Round 1 pooled validation against MC median", round1_val)
    show_df("Round 1 fit closeness summary", round1_close_df)

    plot_round_mc_only(
        "round1",
        ROUND_CONFIGS["round1"]["exposure"],
        round_data["round1"]["obs_stats"],
        PLOT_OUT / "03_round1_mc_only.png"
    )
    plot_round_ode_only(
        "round1",
        ROUND_CONFIGS["round1"]["exposure"],
        round1_curve,
        "Full 5-parameter fit",
        PLOT_OUT / "04_round1_ode_only.png"
    )

    # Round 2 and Round 3 beta-only fits
    beta_only_rounds = ["round2", "round3"]
    beta_init = round1_params[0]

    for round_name in beta_only_rounds:
        if len(round_data[round_name]["pooled_runs"]) == 0:
            continue

        params_r, fit_obj_r, fit_df_r = fit_beta_only(
            round_name,
            round_data[round_name]["pooled_runs"],
            base_params=round1_params,
            beta_init=beta_init
        )

        beta_init = params_r[0]
        fitted_params_by_round[round_name] = params_r.copy()
        fit_tables.append(fit_df_r)

        save_csv(fit_df_r, CSV_OUT / f"10_{round_name}_beta_only_fit_parameters.csv", index=False)

        diag_r = per_run_diagnostics(
            params_r,
            round_data[round_name]["pooled_runs"],
            fit_mode="beta_only_fit_from_round1_baseline"
        )
        diag_tables.append(diag_r)

        curve_r = simulate_curve_dict(params_r, COMMON_GRID)
        save_csv(pd.DataFrame({"t": COMMON_GRID, **curve_r}), CSV_OUT / f"11_{round_name}_fitted_curve.csv", index=False)

        val_r = pooled_curve_validation(
            round_name,
            ROUND_CONFIGS[round_name]["exposure"],
            round_data[round_name]["obs_stats"],
            curve_r,
            fit_mode="beta_only_fit_from_round1_baseline"
        )
        curve_validation_tables.append(val_r)

        close_r = fit_closeness_summary(
            round_name,
            ROUND_CONFIGS[round_name]["exposure"],
            round_data[round_name]["obs_stats"],
            curve_r,
            "beta_only_fit_from_round1_baseline"
        )
        closeness_tables.append(close_r)
        save_csv(close_r, CSV_OUT / f"12_{round_name}_fit_closeness_summary.csv", index=False)

        show_df(f"{safe_round_label(round_name)} beta-only fit", fit_df_r)
        show_df(f"{safe_round_label(round_name)} pooled validation against MC median", val_r)
        show_df(f"{safe_round_label(round_name)} fit closeness summary", close_r)

        plot_round_mc_only(
            round_name,
            ROUND_CONFIGS[round_name]["exposure"],
            round_data[round_name]["obs_stats"],
            PLOT_OUT / f"13_{round_name}_mc_only.png"
        )
        plot_round_ode_only(
            round_name,
            ROUND_CONFIGS[round_name]["exposure"],
            curve_r,
            "Beta-only fit with Round 1 non-beta parameters fixed",
            PLOT_OUT / f"14_{round_name}_ode_only.png"
        )

    fit_all_df = pd.concat(fit_tables, ignore_index=True)
    save_csv(
        fit_all_df,
        CSV_OUT / "15_all_fit_tables_combined.csv",
        index=False,
        preserve_precision=True
    )

    diag_all_df = pd.concat(diag_tables, ignore_index=True)
    save_csv(diag_all_df, CSV_OUT / "16_per_run_diagnostics_all_fitted_rounds.csv", index=False)

    curve_validation_all_df = pd.concat(curve_validation_tables, ignore_index=True)
    save_csv(curve_validation_all_df, CSV_OUT / "17_pooled_curve_validation_all_fitted_rounds.csv", index=False)

    closeness_all_df = pd.concat(closeness_tables, ignore_index=True)
    save_csv(closeness_all_df, CSV_OUT / "18_fit_closeness_summary_all_fitted_rounds.csv", index=False)

    # Predict Round 4 beta
    beta_rounds_available = [r for r in CALIBRATION_ROUNDS if r in fitted_params_by_round]
    if len(beta_rounds_available) < 3:
        raise RuntimeError("Need fitted Round 1-3 beta values to predict Round 4 beta.")

    beta_exposures = np.array([ROUND_CONFIGS[r]["exposure"] for r in beta_rounds_available], dtype=float)
    beta_values = np.array([fitted_params_by_round[r][0] for r in beta_rounds_available], dtype=float)

    beta_slope, beta_intercept = np.polyfit(beta_exposures, beta_values, 1)
    beta4_pred = beta_intercept + beta_slope * ROUND_CONFIGS["round4"]["exposure"]
    beta4_pred = float(np.clip(beta4_pred, BETA_LOWER, BETA_UPPER))

    round4_pred_params = round1_params.copy()
    round4_pred_params[0] = beta4_pred

    beta_prediction_df = pd.DataFrame([{
        "target_round": "round4",
        "target_exposure": ROUND_CONFIGS["round4"]["exposure"],
        "prediction_method": "linear_regression_beta_vs_exposure_using_round1_to_round3",
        "slope": beta_slope,
        "intercept": beta_intercept,
        "predicted_beta_round4": beta4_pred,
        "round1_beta": fitted_params_by_round["round1"][0],
        "round2_beta": fitted_params_by_round["round2"][0],
        "round3_beta": fitted_params_by_round["round3"][0],
        "epsilon_fixed_from_round1": round1_params[1],
        "gamma_fixed_from_round1": round1_params[2],
        "rho_fixed_from_round1": round1_params[3],
        "alpha_fixed_from_round1": round1_params[4],
    }])
    save_csv(
        beta_prediction_df,
        CSV_OUT / "19_round4_beta_prediction.csv",
        index=False,
        preserve_precision=True
    )

    show_df("Predicted Round 4 beta from Rounds 1-3", beta_prediction_df)

    # Round 4 fit and comparison
    beta4_fit_value = np.nan

    if len(round_data["round4"]["pooled_runs"]) > 0:
        round4_fit_params, round4_fit_obj, round4_fit_df = fit_beta_only(
            "round4",
            round_data["round4"]["pooled_runs"],
            base_params=round1_params,
            beta_init=beta4_pred
        )

        beta4_fit_value = round4_fit_params[0]
        fitted_params_by_round["round4"] = round4_fit_params.copy()
        save_csv(round4_fit_df, CSV_OUT / "20_round4_beta_only_fit_parameters.csv", index=False)

        round4_diag = per_run_diagnostics(
            round4_fit_params,
            round_data["round4"]["pooled_runs"],
            fit_mode="beta_only_fit_from_round1_baseline"
        )
        save_csv(round4_diag, CSV_OUT / "21_round4_per_run_diagnostics.csv", index=False)

        round4_obs_stats = build_observed_stats(round_data["round4"]["pooled_runs"])
        round_data["round4"]["obs_stats"] = round4_obs_stats

        round4_pred_curve = simulate_curve_dict(round4_pred_params, COMMON_GRID)
        round4_fit_curve = simulate_curve_dict(round4_fit_params, COMMON_GRID)

        save_csv(
            pd.DataFrame({"t": COMMON_GRID, **round4_pred_curve}),
            CSV_OUT / "22_round4_predicted_curve_from_beta_trend.csv",
            index=False
        )
        save_csv(
            pd.DataFrame({"t": COMMON_GRID, **round4_fit_curve}),
            CSV_OUT / "23_round4_fitted_curve_beta_only.csv",
            index=False
        )

        round4_pred_val = pooled_curve_validation(
            "round4",
            ROUND_CONFIGS["round4"]["exposure"],
            round4_obs_stats,
            round4_pred_curve,
            fit_mode="predicted_beta_from_round1_to_round3_trend"
        )
        round4_fit_val = pooled_curve_validation(
            "round4",
            ROUND_CONFIGS["round4"]["exposure"],
            round4_obs_stats,
            round4_fit_curve,
            fit_mode="beta_only_fit_from_round1_baseline"
        )
        save_csv(
            pd.concat([round4_pred_val, round4_fit_val], ignore_index=True),
            CSV_OUT / "24_round4_prediction_vs_fit_validation.csv",
            index=False
        )

        round4_pred_close_df = fit_closeness_summary(
            "round4",
            ROUND_CONFIGS["round4"]["exposure"],
            round4_obs_stats,
            round4_pred_curve,
            "predicted_beta_from_round1_to_round3_trend"
        )
        save_csv(round4_pred_close_df, CSV_OUT / "25_round4_predicted_fit_closeness_summary.csv", index=False)

        round4_fit_close_df = fit_closeness_summary(
            "round4",
            ROUND_CONFIGS["round4"]["exposure"],
            round4_obs_stats,
            round4_fit_curve,
            "beta_only_fit_from_round1_baseline"
        )
        save_csv(round4_fit_close_df, CSV_OUT / "26_round4_fitted_fit_closeness_summary.csv", index=False)

        round4_compare_df = pd.DataFrame([{
            "round": "round4",
            "exposure": ROUND_CONFIGS["round4"]["exposure"],
            "predicted_beta": beta4_pred,
            "fitted_beta": beta4_fit_value,
            "absolute_error": abs(beta4_pred - beta4_fit_value),
            "relative_error_percent": (
                abs(beta4_pred - beta4_fit_value) / abs(beta4_fit_value) * 100.0
                if beta4_fit_value != 0 else np.nan
            ),
        }])
        save_csv(round4_compare_df, CSV_OUT / "27_round4_predicted_vs_fitted_beta.csv", index=False)

        show_df("Round 4 beta-only fit", round4_fit_df)
        show_df("Round 4 predicted-curve validation", round4_pred_val)
        show_df("Round 4 fitted-curve validation", round4_fit_val)
        show_df("Round 4 predicted-curve closeness summary", round4_pred_close_df)
        show_df("Round 4 fitted-curve closeness summary", round4_fit_close_df)
        show_df("Round 4 predicted beta vs fitted beta", round4_compare_df)

        plot_round_mc_only(
            "round4",
            ROUND_CONFIGS["round4"]["exposure"],
            round4_obs_stats,
            PLOT_OUT / "16_round4_mc_only.png"
        )

        plot_round_ode_only(
            "round4",
            ROUND_CONFIGS["round4"]["exposure"],
            round4_pred_curve,
            "Predicted ODE using beta-exposure trend",
            PLOT_OUT / "17_round4_predicted_ode_only.png"
        )

        plot_round_ode_only(
            "round4",
            ROUND_CONFIGS["round4"]["exposure"],
            round4_fit_curve,
            "Fitted ODE using beta-only fit",
            PLOT_OUT / "18_round4_fitted_ode_only.png"
        )

    plot_beta_exposure_trend(
        exposures=beta_exposures,
        betas=beta_values,
        beta4_pred=beta4_pred,
        beta4_fit=beta4_fit_value if np.isfinite(beta4_fit_value) else None,
        outpath=PLOT_OUT / "19_beta_vs_exposure_trend.png"
    )

    # ODE vs MC infected AUC distribution comparison
    ode_vs_mc_auc_df = build_ode_vs_mc_auc_distribution_summary(
        round_data,
        fitted_params_by_round
    )
    save_csv(
        ode_vs_mc_auc_df,
        CSV_OUT / "28_ode_vs_mc_auc_distribution_summary.csv",
        index=False
    )
    show_df("ODE vs MC infected AUC distribution summary", ode_vs_mc_auc_df)

    plot_ode_vs_mc_auc_distribution(
        round_data,
        fitted_params_by_round,
        PLOT_OUT / "20_ode_vs_mc_auc_distribution_boxplot.png"
    )

    show_section("Workflow completed")
    print(f"{CITY_NAME} outputs saved to:\n{OUTPUT_DIR}")
    print(f"CSV folder: {CSV_OUT}")
    print(f"Plot folder: {PLOT_OUT}")


if __name__ == "__main__":
    main()
```
